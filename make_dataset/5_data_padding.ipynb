```python
# 使用するライブラリのインポート
from keras_preprocessing.image import load_img, img_to_array
from keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
import numpy as np
import os
import glob
```

```python
# 水増しのプログラム
#https://algorithm.joho.info/programming/python/flask-keras-image-extend/ を参考にした

# 1枚当たり10枚の画像を水増し
PADDING_RATE = 10

# 入力画像の保存先パス
input_path = "/content/drive/Shareddrives/第5回中高生情報学研究コンテスト/手書き数字のデータセット/学習用画像"
folder = ["dataset.0", "dataset.1", "dataset.2", "dataset.3", "dataset.4", 
          "dataset.5", "dataset.6", "dataset.7", "dataset.8", "dataset.9"]

# 出力画像の保存パス
output_path = "/content/drive/Shareddrives/第5回中高生情報学研究コンテスト/水増しした手書き数字のデータセット/学習用画像"
if os.path.isdir(output_path) == False:
  os.mkdir(output_path)

 # ImageDataGeneratorの生成
datagen = ImageDataGenerator(rotation_range = 20, # ランダムに回転させる範囲
                             width_shift_range = 0.2, # ランダムに幅をシフトさせる範囲
                             height_shift_range = 0.2, # ランダムに高さをシフトさせる範囲
                             zoom_range=0.2 # ランダムにズームさせる範囲
)


for folder_name in folder:
  read_data = input_path + '/' + folder_name
  jpg_files = glob.glob(read_data + "/*.jpg")
  output_data_path = output_path + '/' + folder_name
  if os.path.isdir(output_data_path) == False:
    os.mkdir(output_data_path)
  
  # ここから水増しする
  for jpg_file in jpg_files:
    img = load_img(jpg_file)
    x = img_to_array(img)
    x = np.expand_dims(x, axis=0)
    dg = datagen.flow(x, batch_size=1, save_to_dir=output_data_path, save_prefix="img", save_format="jpg")
    for i in range(PADDING_RATE):
      batch = dg.next()
```
